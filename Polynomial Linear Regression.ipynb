```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/Position_Salaries.csv')
dataset
```

```python
x= dataset.iloc[:, [1]].values
x              
```

```python
y = dataset.iloc[:, -1].values
y
```

##Simple Linear Reg Model

```python
from sklearn.linear_model import LinearRegression
S_regressor = LinearRegression()
S_regressor.fit(x, y)
```

```python
S_regressor.predict([[6.5]])
```

##Polynomial Linear Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4)
```

Here degree is equals to the value of power of x

```python
x_poly = poly.fit_transform(x)
P_regressor = LinearRegression()
P_regressor.fit(x_poly, y)
```

```python
P_regressor.predict(poly.fit_transform([[6.5]]))
```

##Visualization Of SLR Prediction

```python
plt.scatter(x, y, color='red', marker='*')
plt.plot(x, S_regressor.predict(x))
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

##Visualization Of PLR Prediction

```python
plt.scatter(x, y, color='red', marker='*')
plt.plot(x, P_regressor.predict(x_poly), color='green')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
